```python
iris_df = pd.read_csv("iris.csv", header=1)
iris_df.head()
```

```python
new_names =  {'5.1': 'sepal_length',
               '3.5': 'sepal_width',
               '1.4': 'petal_length',
               '0.2': 'petal_width',
               'setosa': 'species'}
```

```python
iris_df.rename(columns=new_names, inplace=True)
iris_df.head()
```

```python
x=iris_df['sepal_length'].loc[0:20]
y=iris_df['sepal_width'].loc[0:20]

plt.pie(y, labels=x, autopct='%1.1f%%') 
plt.show()
```

```python
x=iris_df['sepal_length'].loc[0:20]
y=iris_df['sepal_width'].loc[0:20]

x_coords = np.arange(len(x)) 
plt.bar(x_coords, y) 
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
attributs = ('sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species') 
jumlah = (150, 150, 150, 150, 150) 
df = pd.DataFrame({ 'attribut': attributs, 'jumlah': jumlah, }) 
df.sort_values(by='jumlah', inplace=True) 
x_coords = np.arange(len(df)) 
plt.bar(x_coords, df['jumlah'], tick_label=df['attribut']) 
plt.xticks(rotation=90) 
plt.ylabel('Jumlah') 
plt.title('Data Iris') 
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
attributs = ('sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species') 
jumlah = (150, 150, 150, 150, 150) 
df = pd.DataFrame({ 'attribut': attributs, 'jumlah': jumlah, }) 
df.sort_values(by='jumlah', inplace=True) 
x_coords = np.arange(len(df)) 
colors = ['#0000FF' for _ in range(len(df))] 
colors[-1] = '#FF0000'  
plt.bar(x_coords, df['jumlah'], tick_label=df['attribut'], color=colors) 
plt.xticks(rotation=90) 
plt.ylabel('Jumlah') 
plt.title('Data Iris') 
plt.show()
```
